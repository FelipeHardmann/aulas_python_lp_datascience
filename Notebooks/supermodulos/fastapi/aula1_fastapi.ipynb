###O que são API's e qual é sua importância para o desenvolvimento de sistemas?

Uma API (Application Programming Interface) é um conjunto de regras e definições que permite a comunicação entre diferentes sistemas, aplicações ou serviços. Ela funciona como uma ponte que define como os programas podem interagir entre si, seja para enviar ou receber dados.


###Principais características de uma API:

Interoperabilidade: Permite que diferentes sistemas e linguagens de programação se comuniquem.

Padronização: Define regras para as requisições e respostas (geralmente seguindo padrões como REST ou GraphQL).

Abstração: Esconde a complexidade do funcionamento interno de um sistema e fornece apenas o que é necessário para o uso externo.

###Tipos de API's

RESTful API: Baseada em HTTP e usa métodos como GET, POST, PUT, DELETE.

GraphQL: Permite requisições mais flexíveis, onde o cliente define exatamente os dados que deseja receber.

SOAP: Um protocolo mais antigo que utiliza XML para comunicação.

WebSocket API: Mantém uma conexão aberta para comunicação bidirecional em tempo real.

Nas aulas desse curso iremos nos aprofundar sobre o conceito de REST, por ser o padrão mais utilizado no mercado.

A maior parte das API's encontradas na internet utilizam o conceito de Rest, que é Representational State Transfer.

REST (ou RESTful) é um conjunto de restrições de arquitetura, e não um protocolo ou padrão. Os desenvolvedores de API podem implementar o REST de diversas maneiras.

Quando uma solicitação de cliente é feita por uma API RESTful, ela transfere uma representação do estado do recurso para o solicitante ou endpoint. Essa informação, ou representação, é entregue em um de vários formatos, via HTTP: JSON (Javascript Object Notation), HTML, XLT, Python, PHP ou texto simples. O JSON é o mais usado porque, apesar de seu nome, ele é um formato independente de linguagem e pode ser lido por máquinas e humanos.

Algo importante a considerar: headers e parâmetros também são essenciais nos métodos HTTP de uma API RESTful por conterem identificadores fundamentais para os metadados da requisição, autorização, URI, cache, cookies e outros Os cabeçalhos de solicitação e resposta têm suas próprias informações de conexão HTTP e códigos de status.

O criador do Rest é Roy Fielding, que ajudou na criação do protocolo HTTP, que é onde opera a internet, isso auxília o fato do REST caminhar junto com o protocolo HTTP, pois, o REST foi projetado para aproveitar ao máximo as características do HTTP de forma eficiente e padronizada, se baseando sempre nos métodos HTTP.

##Concorrência

Na ciência da computação, a concorrência é a execução de múltiplas intruções sequencias ao mesmo tempo.

Na programação existem dois tipos de concorrência:

- Programação Paralela
- Programação Assíncrona


Quando utilizar programação paralela?

Usamos a programação paralela quando utilizamos tarefas que fazem uso excessivo da CPU.

Exemplos:

Operações de strings, algoritmos de busca, processamentos númericos.

Quando utilizar programação assíncrona?

A programação assíncrona é melhor quando exige o uso do IO.

Exemplos:

Leitura e escrita em bancos de dados, chamadas de API's, Cópia, upload de dados.

Ou seja, tudo aquilo que envolve entrada e saída de dados.

##Parâmetros da rota da URL

Dentro do FastAPI você pode declarar os "parâmetros" ou "variáveis" com a mesma sintaxe utilizada pelo formato de strings do Python:

```
from fastapi import FastAPI

app = FastAPI()


@app.get("/nomes/{nome}")
async def nome(nome):
    return {"message": nome}
```

###Documentação do FastAPI

As Vantagens da Documentação Automática do FastAPI

O FastAPI se destaca não apenas por sua performance e facilidade de uso, mas também por sua documentação automática e interativa, que é uma das melhores entre os frameworks Python. Essa funcionalidade é possível graças à integração com o OpenAPI (antigo Swagger) e o JSON Schema, gerando uma interface intuitiva para testar e entender sua API sem esforço adicional.

📌 Principais Vantagens da Documentação do FastAPI

1️⃣ Documentação Automática e Atualizada

    O FastAPI gera automaticamente a documentação da sua API com base nos modelos (Pydantic) e rotas definidas.

    Sem necessidade de escrever manualmente:

        Endpoints

        Parâmetros (Query, Path, Body)

        Modelos de requisição/resposta

        Códigos de status HTTP

    Se você alterar um modelo ou rota, a documentação se atualiza instantaneamente.

2️⃣ Interface Interativa (Swagger UI e ReDoc)

Por padrão, o FastAPI disponibiliza duas interfaces de documentação:
🔹 Swagger UI (/docs)

    Permite testar a API diretamente no navegador.

    Mostra exemplos de requisições e respostas.

    Suporte a autenticação (JWT, OAuth2).

    Facilita o debug durante o desenvolvimento.

🔹 ReDoc (/redoc)

    Focado em legibilidade e organização.

    Exibe a documentação em um formato mais limpo e estruturado.

    Ideal para compartilhar com a equipe ou clientes.

3️⃣ Validação e Exemplos Incorporados

    Os schemas do Pydantic garantem que a documentação mostre:

        Tipos de dados esperados (ex.: str, int, List[Item]).

        Restrições (ex.: min_length, max_length, regex).

        Exemplos de payloads (útil para front-end e integrações).

4️⃣ Suporte a OpenAPI e JSON Schema

    A documentação segue o padrão OpenAPI 3.1, permitindo:

        Integração com ferramentas como Postman, Insomnia.

        Geração de clientes SDK automaticamente.

        Compatibilidade com ferramentas de monitoramento (ex.: Prometheus, Grafana).

5️⃣ Economia de Tempo e Menos Erros

    Evita discrepâncias entre código e documentação.

    Reduz a necessidade de manuais e tutoriais separados.

    Facilita a onboarding de novos desenvolvedores.
