# Introdução às Fórmulas DAX

## O que são Fórmulas DAX?

DAX (Data Analysis Expressions) é uma linguagem de fórmula utilizada no Power BI, Power Pivot no Excel e no SSAS (SQL Server Analysis Services) Tabular. Ela é projetada para trabalhar com dados relacionais e realizar cálculos complexos em tabelas e colunas.

## Para que Utilizar as Fórmulas DAX?

As fórmulas DAX são utilizadas para:

1. **Criação de Colunas Calculadas**: Você pode criar novas colunas em suas tabelas com base em cálculos que envolvem outras colunas.
2. **Medidas**: São cálculos dinâmicos que podem ser usados em visualizações, como somas, médias, contagens, etc.
3. **Tabelas Calculadas**: Permitem criar novas tabelas com base em cálculos ou filtros aplicados a tabelas existentes.
4. **Filtros e Contextos**: DAX permite trabalhar com contextos de filtro, o que é essencial para análises detalhadas e segmentadas.

## Quando Utilizar as Fórmulas DAX?

Você deve utilizar fórmulas DAX quando:

- Precisa realizar cálculos que não podem ser feitos diretamente com as ferramentas de transformação de dados (Power Query).
- Deseja criar métricas complexas que dependem de contextos de filtro.
- Precisa de cálculos dinâmicos que se ajustam conforme os filtros aplicados pelos usuários.

## Importância das Fórmulas DAX no Power BI

As fórmulas DAX são fundamentais no Power BI porque:

- **Permitem Análises Dinâmicas**: Com DAX, você pode criar medidas que se ajustam conforme os filtros aplicados, proporcionando uma análise mais flexível e poderosa.
- **Aumentam a Capacidade de Análise**: Com DAX, você pode ir além das agregações simples, criando cálculos complexos que atendem às necessidades específicas do negócio.
- **Integração com Outras Ferramentas**: DAX é compatível com outras ferramentas da Microsoft, como Excel e SSAS, o que facilita a integração e a consistência dos dados.

## Como Utilizar as Fórmulas DAX no Power BI

1. **Criando uma Medida**:
   - No Power BI, vá para a aba "Modelagem" e clique em "Nova Medida".
   - Escreva a fórmula DAX. Por exemplo, para calcular o total de vendas:
     ```dax
     Total Vendas = SUM(TabelaVendas[ValorVenda])
     ```

2. **Criando uma Coluna Calculada**:
   - No Power BI, vá para a aba "Modelagem" e clique em "Nova Coluna".
   - Escreva a fórmula DAX. Por exemplo, para calcular o lucro:
     ```dax
     Lucro = TabelaVendas[ValorVenda] - TabelaVendas[Custo]
     ```

3. **Criando uma Tabela Calculada**:
   - No Power BI, vá para a aba "Modelagem" e clique em "Nova Tabela".
   - Escreva a fórmula DAX. Por exemplo, para criar uma tabela de vendas filtrada:
     ```dax
     Vendas Filtradas = FILTER(TabelaVendas, TabelaVendas[ValorVenda] > 1000)
     ```
