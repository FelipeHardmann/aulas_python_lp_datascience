###Introdução a relatórios e gráficos no Power Bi.

Antes de começarmos a ver os relatórios e gráficos, vamos entender um pouco mais para que serve cada tipo de gráfico

### 1. **Gráficos de Colunas e Barras**

**Caso de Uso**: Ideal para comparar valores entre diferentes categorias ou períodos, como vendas mensais por região ou total de funcionários por departamento.

- Coluna: Exibe dados em barras verticais, útil para comparar valores entre diferentes categorias.

- Barras: Similar ao gráfico de colunas, mas as barras são horizontais. Ideal para mostrar valores longos ou muitas categorias.

### 2. **Gráfico de Linhas**

**Caso de Uso**: Análises temporais, como acompanhar tendências de vendas semanais ou evolução da receita ao longo do tempo.

- Exibe informações sequenciais ao longo do tempo, ótimo para mostrar tendências e mudanças.

### 3. **Gráfico de Área**

**Caso de Uso**: Útil para mostrar a contribuição de diferentes partes em um total ao longo do tempo, como uma análise cumulativa de crescimento.

- Semelhante ao gráfico de linhas, mas com áreas preenchidas. Indicado para enfatizar o volume total de uma tendência ao longo do tempo.

### 4. **Gráfico de Pizza e Donut**

**Caso de Uso**: Representação de partes de um todo, como distribuição de market share ou percentual de cada setor em uma carteira de investimentos.

- Pizza: Útil para mostrar proporções de uma única categoria.

- Donut: Uma variação da pizza, com o centro vazio, que também permite comparar partes de um todo.

### 5. **Gráfico de Dispersão**

**Caso de Uso**: Análises de correlação, como entender a relação entre orçamento de marketing e vendas ou comparar volume de vendas e lucros por produto.

- Exibe a relação entre duas variáveis. Cada ponto representa um par de valores, útil para identificar padrões, tendências e correlações.

### 6. **Mapa (Mapas Básicos e Mapas de Calor)**

**Caso de Uso**: Ideal para dados com localização geográfica, como vendas por país, cidade, ou estado, ou para exibir concentração de dados (ex.: calor em uma cidade com mais clientes)

- Mapa Básico: Mostra a distribuição geográfica de dados.

- Mapa de Calor: Indica intensidade ou concentração de dados em diferentes áreas geográficas.

### 7. **Tabela e Matriz**

**Caso de Uso**: Boas para exibir detalhes de dados com múltiplas categorias, como relatórios detalhados de vendas ou dados financeiros.

- Tabela: Exibe dados tabulares com colunas e linhas.

- Matriz: Tabela interativa com agrupamento hierárquico, permitindo expandir e recolher níveis de dados.

### 8. **Cartões e Cartões com KPI**

**Caso de Uso**: Útil para destacar KPIs (indicadores-chave de desempenho), como receita total, lucro bruto, ou total de clientes

- Cartão: Mostra valores únicos, como totais ou média.

- Cartão com KPI: Apresenta uma métrica de desempenho com comparações a uma meta.

### 9. **Segmentação de Dados (Slicers)**

**Caso de Uso**: Controle fácil para o usuário interagir com dados, como segmentar relatórios por região, ano, ou categoria.

- Filtro visual interativo que permite segmentar os dados por categoria ou valor.

### 10. **Gráfico de Funil**

**Caso de Uso**: Acompanhar processos que reduzem dados ao longo de estágios, como o funil de vendas ou de conversão de leads.

- Mostra um processo com múltiplas etapas, como funil de vendas ou etapas de um fluxo.

13. Gráfico de Combo (Linha e Coluna)

- Combina gráficos de linha e coluna em um só, útil para exibir dados com duas variáveis
