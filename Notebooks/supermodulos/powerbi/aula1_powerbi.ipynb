# Introdução ao Power BI

O **Power BI** é uma ferramenta de **análise de dados e visualização** desenvolvida pela **Microsoft**. Seu principal objetivo é transformar dados brutos em **relatórios interativos**, **dashboards** e **gráficos**, permitindo uma análise mais estratégica e eficiente para tomada de decisões empresariais.


## Para que serve o Power BI?

O Power BI é uma ferramenta projetada para:
- **Analisar dados** de maneira interativa.
- **Criar relatórios dinâmicos** e **gráficos interativos**.
- **Ajudar na tomada de decisões estratégicas**, a partir de informações claras e visuais.
- **Facilitar a comunicação de insights** através de dashboards e relatórios acessíveis.

Ele é ideal para empresas de todos os portes que desejam transformar dados em **informações estratégicas**, além de oferecer ferramentas para integração com diversas fontes de dados (Excel, SQL, Google Analytics, etc.).


## Quando utilizar o Power BI?

Você deve utilizar o Power BI quando precisar:
- **Analisar grandes volumes de dados** de forma rápida e eficaz.
- Criar **relatórios dinâmicos** que se atualizam em tempo real.
- **Explorar dados** de diferentes fontes e transformá-los em **gráficos interativos** e **dashboards**.
- **Tomar decisões baseadas em dados** em tempo real.
- Facilitar a **visualização** e a **comunicação de dados** com outros membros da equipe ou stakeholders.


## Por que foi criado?

O Power BI foi desenvolvido pela Microsoft com o objetivo de **simplificar o processo de análise de dados**, permitindo que **usuários de negócios** (não apenas especialistas em TI) possam explorar e analisar dados de forma prática e eficiente. Seu foco principal é facilitar a **tomada de decisões baseadas em dados concretos** e apresentados de maneira visual e clara.

## Quem criou o Power BI?

O Power BI foi criado pela **Microsoft** e lançado em 2014. A Microsoft desenvolveu o Power BI com o objetivo de integrar suas ferramentas ao crescente mercado de **Business Intelligence** e análise de dados, fornecendo uma solução acessível e poderosa para empresas de diversos segmentos.


## Ele substitui quem?

O Power BI pode substituir diversas ferramentas de **Business Intelligence (BI)** tradicionais, como:
- **Excel** (quando se busca mais recursos de visualização e análise de dados dinâmicos).
- **Tableau** (concorrente direto na criação de dashboards e relatórios interativos).
- **QlikView** (outras ferramentas de BI que também possuem foco em visualização e exploração de dados).

A principal vantagem do Power BI está na **integração** com outras ferramentas da Microsoft, como o **Excel**, **Azure**, e a **plataforma de nuvem** da empresa, além de **atualizações automáticas de dados** e maior **performance** ao lidar com grandes volumes de informação.

### Conhecendo um pouco mais da ferramenta

Quando vemos a tela do Power Bi, temos 3 guias básicas que utilizamos normalmente para fazer a nossa manipulação de dados, que são elas **Relatório**, **Tabela** e **Modelo**.

- 1. Relatório: Nessa guia vamos construir os gráficos, dashboards, relatórios.
- 2. Tabela: Essa guia é muito similar ao Excel, normalmente vamos ver a exibição dos dados como uma planilha similar ao Excel.
- 3. Modelo: Nessa guia é onde podemos cruzar informações de tabelas diferentes para exibição em um mesmo relatório ou seja, criaremos relacionamentos entre as tabelas.

image.png

Agora, no Power Bi mais atual, temos também um novo campo chamado Dax, esse Dax é basicamente uma coleção de funções, operadores e constantes que podem ser usados em uma fórmula ou expressão, para calcular e retornar um ou mais valores. Para falar de forma mais simples, o DAX ajuda a criar novas informações de dados que já estão em seu modelo. O DAX agora, está acompanhado do Copilot que te ajuda a fazer algumas dessas manipulações.

Outro problema que estamos enfrentando nessa tabela é que ao importarmos a tabela, ele não entendeu que a nossa primeira linha seria o Cabeçalho das nossas colunas.

Podemos ajustar isso de forma muito fácil no Power Query, somente clicando no campos "Usar a Primeira Linha como Cabeçalho". Nisso, o Power Query já ajusta nossa primeira linha como cabeçalho da nossa planilha, nesse caso, se você estiver prestando atenção, uma coluna mudou até o ícone que agora está com números.

Isso indica que essa coluna agora armazena somente números.


Mas, com essa alteração, também, vemos um pequeno problema na coluna de Contratação, nesse caso, essa coluna está com um ABC, que indica que é uma coluna de texto, mas, nesse caso, não queremos uma coluna de texto e sim um coluna do tipo de data, mas, como a formatação da data está de forma diferente, o Power Bi, não consegue entender que aquilo é uma data e é muito importante que aplicarmos os tipos de maneiras corretas no Power Bi, pois isso facilita as manipulações utilizando aquele tipo.

Para alterarmos isso, basta clicarmos com o botão direito em cima da coluna, que queremos mudar esse tipo e irmos na opção "Alterar Tipo", nessa opção, podemos fazer a alteração para o tipo de data de forma fácil, porém, essas datas, estão no padrão americano, que não é comum por aqui(BR), nesse caso, precisamos que seja dia/mês/ano, portanto, quando clicarmos em "**Alterar tipo**, vamos na última opção, que é "**Usando a localidade**", que nesse caso, podemos fazer alteração informando que o tipo que queremos é o data e que o tipo de formatação atual é de data americana, agora, o Power Query vai aplicar as transformações necessárias e trazer a nossa coluna para a tipagem correta.
