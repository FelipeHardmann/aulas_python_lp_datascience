##Modelagem de dados

Na aula de hoje, vamos aprender um pouco, sobre modelagem de dados.

**Lembre-se, esse é um dos conceitos mais importantes da análise de dados com Power Bi.**

A **modelagem de dados no Power BI** é essencial para garantir a eficiência, precisão e desempenho dos relatórios e dashboards. Ela permite organizar, estruturar e relacionar diferentes fontes de dados, facilitando análises mais profundas e evitando redundâncias.  

Com uma boa modelagem, os cálculos e medidas DAX ficam mais otimizados, os relatórios carregam mais rapidamente e os insights se tornam mais precisos. Além disso, a criação de relacionamentos entre tabelas possibilita a construção de visualizações dinâmicas e interativas, melhorando a experiência do usuário.  

Dominar a modelagem de dados no Power BI é fundamental para transformar dados brutos em informações estratégicas e acionáveis.

**Tabela Fato**

A tabela fato armazena os eventos ou transações que ocorrem ao longo do tempo, geralmente essas tabelas armazenam mais valores, geralmente são bem grandes.

**Tabela Dimensão**

As tabelas dimensão armazenam informações descritivas relacionadas à tabela fato. Elas ajudam a filtrar, categorizar e segmentar os dados.

É muito importante entendermos esse tipo de separação, por conta da perfomance do modelo de dados e por facilitar análises mais detalhadas e deshboards totalmente interativos.
