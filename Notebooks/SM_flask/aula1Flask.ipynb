##Como a Web funciona

###HTTP

O protocolo de comunicação HTTP pode ser considerado como a espinha dorsal da web moderna. Para qualquer desenvolvedor que trabalha com a internet, um entendimento sólido de HTTP é essencial. Ele permite a criação de aplicações web eficientes, seguras e interoperáveis, além de fornecer a base para a comunicação e troca de dados na internet.

### O que é o HTTP

HTTP (Hypertext Transfer Protocol) é um protocolo de comunicação utilizado na transferência de informações na web. É a base para qualquer troca de dados na internet, permitindo que navegadores da web e servidores se comuniquem entre si.

Então, em resumo o cliente(que nesse caso não precisa ser necessariamente o usuário) faz uma requisição para algum servidor e esse servidor fornece uma resposta para esse cliente

Em resumo, esse protocolo é uma comunicação entre cliente e servidor

###Requisição

Como vimos no tópico anterior, o HTTP é um estilo de comunicação entre o cliente e servidor. Entretanto, essa comunicação não acontece de qualquer forma, por isso o nome protocolo HTTP, porque há uma série de regras para fazer a requisição no servidor, já que, mais de um cliente pode fazer a requisição ao mesmo tempo.

Essa requisição acontece com várias coisas, por exemplo, o cliente pode fazer a requisição de um HTML, um JPG, entre outros tipos de arquivos.

###Verbos HTTP
Os verbos HTTP, também conhecidos como métodos HTTP, são usados para especificar a ação desejada a ser executada em um recurso especificado. Aqui estão os principais verbos HTTP:

1. **GET**:
   - **Uso**: Recuperar dados de um servidor.
   - **Características**:
     - É uma operação de leitura, ou seja, não altera o estado do recurso no servidor.
     - Pode ser armazenado em cache.
     - Pode ser armazenado em histórico do navegador.
   - **Exemplo**: `GET /api/usuarios` (Recupera a lista de usuários)

2. **POST**:
   - **Uso**: Enviar dados ao servidor para criar um novo recurso.
   - **Características**:
     - Envia dados no corpo da solicitação.
     - Não é idempotente, ou seja, múltiplas requisições podem causar múltiplos efeitos.
   - **Exemplo**: `POST /api/usuarios` (Cria um novo usuário)

3. **PUT**:
   - **Uso**: Atualizar um recurso existente ou criar um recurso se ele não existir.
   - **Características**:
     - Envia dados no corpo da solicitação.
     - É idempotente, ou seja, múltiplas requisições terão o mesmo efeito que uma única requisição.
   - **Exemplo**: `PUT /api/usuarios/123` (Atualiza o usuário com ID 123)

4. **PATCH**:
   - **Uso**: Aplicar modificações parciais a um recurso existente.
   - **Características**:
     - Envia dados no corpo da solicitação.
     - Não necessariamente idempotente, pois pode depender da implementação.
   - **Exemplo**: `PATCH /api/usuarios/123` (Atualiza parcialmente o usuário com ID 123)

5. **DELETE**:
   - **Uso**: Remover um recurso do servidor.
   - **Características**:
     - É uma operação de exclusão.
     - É idempotente.
   - **Exemplo**: `DELETE /api/usuarios/123` (Remove o usuário com ID 123)

6. **HEAD**:
   - **Uso**: Recuperar os cabeçalhos de resposta, sem o corpo da resposta.
   - **Características**:
     - Similar ao GET, mas sem o corpo da resposta.
     - Útil para verificar se um recurso existe ou obter metadados.
   - **Exemplo**: `HEAD /api/usuarios` (Recupera os cabeçalhos da lista de usuários)

7. **OPTIONS**:
   - **Uso**: Descrever as opções de comunicação para o recurso de destino.
   - **Características**:
     - Útil para verificar os métodos suportados pelo servidor para um recurso específico.
   - **Exemplo**: `OPTIONS /api/usuarios` (Recupera os métodos HTTP suportados para a lista de usuários)

Esses métodos são fundamentais para a comunicação cliente-servidor na web, permitindo operações de criação, leitura, atualização e exclusão (CRUD) de recursos.

###Response

Nesse caso, assim, como há a requisição, temos também a resposta do servidor para o cliente e essa resposta vai depender de uma série de fatores, como, se o cliente tem autorização, se há aquela requisição no servidor e entre outros.

###Status Code
Os status codes (códigos de status) são códigos numéricos retornados por um servidor em resposta a uma solicitação HTTP feita por um cliente (como um navegador web ou uma API client). Eles indicam se a solicitação foi bem-sucedida, se houve algum erro ou se alguma ação adicional é necessária por parte do cliente. Aqui está uma visão geral dos principais grupos de códigos de status HTTP:

1. **1xx: Informacional**:
   - Indica que a solicitação foi recebida e está sendo processada.
   - **Exemplo**:
     - `100 Continue`: O servidor recebeu a solicitação inicial e o cliente pode continuar com o envio do corpo da solicitação.

2. **2xx: Sucesso**:
   - Indica que a solicitação foi recebida, compreendida e aceita com sucesso.
   - **Exemplos**:
     - `200 OK`: A solicitação foi bem-sucedida e o servidor retornou os dados solicitados.
     - `201 Created`: Um novo recurso foi criado como resultado da solicitação.

3. **3xx: Redirecionamento**:
   - Indica que o cliente precisa realizar alguma ação adicional para completar a solicitação.
   - **Exemplos**:
     - `301 Moved Permanently`: O recurso solicitado foi movido permanentemente para uma nova URL.
     - `302 Found`: O recurso solicitado foi encontrado em uma URL diferente temporariamente.

4. **4xx: Erro do Cliente**:
   - Indica que houve um problema com a solicitação feita pelo cliente.
   - **Exemplos**:
     - `400 Bad Request`: A solicitação não pôde ser entendida ou estava incorreta.
     - `401 Unauthorized`: A solicitação requer autenticação do usuário.
     - `403 Forbidden`: O servidor entendeu a solicitação, mas está recusando-a.
     - `404 Not Found`: O recurso solicitado não foi encontrado no servidor.

5. **5xx: Erro do Servidor**:
   - Indica que o servidor encontrou um erro ao processar a solicitação.
   - **Exemplos**:
     - `500 Internal Server Error`: O servidor encontrou um erro interno ao processar a solicitação.
     - `502 Bad Gateway`: O servidor, atuando como um gateway ou proxy, recebeu uma resposta inválida do servidor upstream.
     - `503 Service Unavailable`: O servidor está temporariamente indisponível, geralmente devido a manutenção ou sobrecarga.

Esses códigos ajudam a diagnosticar problemas de comunicação entre o cliente e o servidor e fornecem informações úteis para depuração e tratamento de erros nas aplicações web.

##Flask

Agora que vimos os principais tópicos, podemos começar a ver o framework e entender sobre o que é o Flask.

###O que é o Flask
O Flask é um microframework web escrito em Python de fácil utilização e de recursos incríveis. Por se tratar de um microframework, o Flask possui apenas o mínimo possível para funcionar. A partir daí tudo que for necessário para o desenvolvimento da aplicação, precisaremos instalar pacotes extras. Isso permite que um projeto escrito com o Flask só possua o que você realmente vá utilizar.

Além disso, o Flask possui um núcleo simples, mas estendível, não possuindo uma camada de abstração do banco de dados, validação de formulários ou qualquer outro componente. Assim, o Flask suporta extensões de bibliotecas externas capazes de adicionar tais funcionalidades na aplicação final, como veremos ao longo de todo o curso.

O Flask é baseado em duas principais bibliotecas: a Werkzeug e a Jinja2.

A Werkzeug é uma biblioteca que possui uma vasta coleção de recursos para trabalhar com requisições HTTP, como:

 - 1 - Análise de cabeçalho HTTP;
  
 - 2 - Facilidade para trabalhar com requisições e respostas;

 - 3 - Depurador interativo;
  
 - 4 - Suporte ao Unicode;
  
 - 5 - Suporte a sessões e cookies;
    
 - 6 - Sistema de roteamento integrado.

Já o Jinja2 é uma biblioteca para trabalhar com templates. Ela possui uma vasta coleção de recursos, como:

  - 1 - Heranças de templates;
  - 2 - Prevenção de XSS;
  - 3 - Sintaxe configurável;
  - 4 - Execução de código Python em páginas HTML.

##Criando nossa primeira aplicação Flask

Para começarmos, devemos criar nosso ambiente virtual, conhecido como o *.venv*. Utilizando o seguinte comando:

*python -m venv .venv*

Ao criarmos nosso ambiente virtual, devemos agora entrar nele, utilizando o seguinte comando.

**LINUX**

*source .venv/bin/activate*

**WINDOWS**

*.venv\Scripts\activate*


###ATENÇÃO:

Em algumas máquinas windows não vamos conseguir ativar o nosso ambiente virtual.

Para resolver basta executar o comando **Set-ExecutionPolicy -Scope Process -ExecutionPolicy Bypass**, porém vale ressaltar que essa solução só funciona para o terminal ao qual o comando foi executado, caso abra um novo terminal o erro irá acontecer novamente e será necessário executar novamente o comando.

Agora, podemos fazer a instalação do Flask, com o comando:

*pip install flask*

Instalado o Flask, agora veremos qual estrutura iremos seguir por padrão nas nossas aplicações, como o mundo Python ainda tem o Django como seu principal framework Python, vamos seguir a mesma estrutura do Django que é o famoso MVT(Model-View-Template).

Para isso, vamos criar as nossas camadas.

Primeiro criando o nosso pacote *app*.

Dentro desse app, vamos criar mais 3 diretórios, que serão elas:

- app

   - models
   - views
   - templates

Somente para entendermos, a camada de models será onde vamos colocar as nossas entidades(tabelas de banco de dados), a camada de templates é onde terá os nossos arquivos HTML para renderizar para o cliente e a camada de views será a camada que faz a comunicação entre o template e a models.

Agora, vamos criar a nossa primeira rota da aplicação, dentro do diretório *views*, vamos criar o arquivo *first_view.py*.

Nesse arquivo, primeiro, vamos chamar a nossa Classe Flask quem vem do módulo flask.

Então, vai ficar:

*from flask import Flask*

Depois que chamarmos a nossa classe Flask, precisamos agora instanciar essa classe, por padrão chamamos ela de app, que foi o nome da nossa aplicação.

*app = Flask(__ __name__ __)*

Esse __ __name__ __ esta sendo representado pelo nome do diretório que vamos chamar no arquivo que vai rodar a aplicação.

Agora, para criarmos a primeira rota na nossa aplicação precisamos chamar da seguinte forma

```
@app.route('/(aqui vamos colocar a rota que quisermos)')
def (nome da função)():
  return 'Mensagem que você deseja colocar'
```

Agora, como devemos dividir a responsabilidade entre arquivos, vamos criar um arquivo *main.py* dentro do nosso diretório *app*.

Nesse arquivo devemos chamar a nossa instância de app, da seguinte forma:

```
from app.views.first_view import app


if __name__ == '__main__':
    app.run()

```

Agora, basta rodar a sua aplicação e verificar se está rodando a sua aplicação.

#Rotas dinâmicas

O Flask permite criar rotas dinâmicas que podem aceitar parâmetros. Por exemplo:

```
@app.route("/ola/<nome>")
def ola(nome):
    return f'Olá {nome}, seja bem-vindo!!'
```
