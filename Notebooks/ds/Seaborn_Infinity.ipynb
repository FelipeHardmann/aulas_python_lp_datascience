```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/infinity/liveIN.xlsx', 'videos')
display(df)
```

#Gráficos de dispersão

##Scatterplot
http://seaborn.pydata.org/generated/seaborn.scatterplot.html#seaborn.scatterplot

```python
fig = sns.scatterplot(data=df, x='Nº de Alunos', y='Nº de Likes')
plt.show(fig)
```

```python
fig = sns.scatterplot(data=df, x='Nº de Alunos', y='Nº de Likes', hue='Responsável')
plt.show(fig)
```

```python
fig = sns.scatterplot(data=df, x='Nº de Alunos', y='Nº de Likes', hue='Responsável', style='Responsável')
plt.show(fig)
```

```python
fig = sns.scatterplot(data=df, x='Nº de Alunos', y='Nº de Likes', hue='Responsável', palette=['pink', 'black'])
plt.show(fig)
```

```python
fig = sns.scatterplot(data=df, x='Nº de Alunos', y='Nº de Likes', hue='Categoria', style='Responsável')
plt.show(fig)
```

#Relplot
http://seaborn.pydata.org/generated/seaborn.relplot.html#seaborn.relplot

Gráfico de disperção relacional


```python
fig2 = sns.relplot(data=df, x='Nº de Alunos', y='Nº de Likes', hue='Categoria', col='Responsável')
plt.show(fig2)
```

```python
fig2 = sns.relplot(data=df, x='Nº de Alunos', y='Nº de Likes', hue='Categoria', col='Responsável')
fig2.set_titles('Reponsável:{col_name}')
plt.show(fig2)
```

#Histogramas

#Displot
http://seaborn.pydata.org/generated/seaborn.displot.html#seaborn.displot

```python
fig = sns.displot(data=df, x='Nº de Alunos')
plt.show(fig)
```

#Histplot
http://seaborn.pydata.org/generated/seaborn.histplot.html#seaborn.histplot

```python
fig = sns.histplot(data=df, x='Nº de Alunos')
plt.show(fig)
```

##Afinal qual a diferença entre o histplot e o displot?
O displot tem como padrão o histplot, porém, ele é mais flexível, ou sejam, podemos mudar o seu padrão com o kind

Verificando os padrões do displot:

```python
sns.displot?
```

```python
fig = sns.displot(data=df, x='Nº de Alunos', kind='kde')
plt.show(fig)
```

###O kde nos traz a densidade, ou seja, onde está mais concentrado a informação

```python
fig = sns.displot(data=df, x='Nº de Alunos', kind='ecdf')
plt.show(fig)
```

```python
sns.displot(data=df, x="Nº de Alunos", kde=True)
```

```python
fig = sns.displot(data=df, x='Nº de Alunos', hue='Responsável')
plt.show(fig)
```

```python
fig = sns.displot(data=df, x='Nº de Alunos', hue='Responsável', col='Responsável')
plt.show(fig)
```

```python
fig = sns.displot(data=df, x='Nº de Alunos', hue='Responsável', col='Responsável', rug=True)
plt.show(fig)
```
