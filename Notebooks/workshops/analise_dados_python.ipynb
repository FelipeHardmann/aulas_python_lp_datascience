### Pandas

O Pandas é uma das principais bibliotecas para análise de dados com Python. Trata-se de uma biblioteca open-source, ou seja, de código aberto, desenvolvida e mantida por uma grande comunidade de desenvolvedores e pesquisadores ao redor do mundo. Por ser amplamente utilizada tanto no meio acadêmico quanto no mercado, o Pandas recebe atualizações constantes, muitas vezes baseadas em problemas reais enfrentados por analistas, cientistas de dados e engenheiros de dados no dia a dia.

Na aula de hoje, focaremos principalmente em um projeto prático utilizando Pandas, com o objetivo de conhecer o poder dessa biblioteca e entender como ela pode ajudar na resolução de problemas que tradicionalmente seriam feitos em planilhas, como no Excel, mas que podem ser realizados de forma muito mais eficiente através de código.

O Pandas foi criado para facilitar a manipulação, limpeza, organização e análise de dados estruturados. Ele trabalha principalmente com duas estruturas de dados muito importantes:

* Series – uma estrutura semelhante a uma coluna de dados.

* DataFrame – uma estrutura semelhante a uma tabela ou planilha, com linhas e colunas.

### Pandas 🤜 X 🤛 Excel

Quando comparamos o Pandas com ferramentas tradicionais como o Excel, algumas vantagens se destacam:

__Automação de tarefas__
No Excel, muitas operações precisam ser feitas manualmente ou com fórmulas. Com Pandas, é possível escrever um script que executa todo o processo automaticamente, economizando tempo e reduzindo erros.

__Reprodutibilidade__
Quando utilizamos código, conseguimos reproduzir exatamente a mesma análise quantas vezes forem necessárias. Em planilhas, muitas vezes as transformações ficam difíceis de rastrear.

__Escalabilidade__
Planilhas possuem limitações quando trabalhamos com grandes volumes de dados. O Pandas consegue lidar com bases muito maiores, permitindo análises que seriam difíceis ou até impossíveis em ferramentas tradicionais.

__Integração com outras ferramentas__
Como faz parte do ecossistema Python, o Pandas pode ser facilmente integrado com outras bibliotecas de dados, como ferramentas de visualização, estatística e machine learning.

__Processos mais organizados e versionados__
Utilizando código, podemos versionar nossas análises com ferramentas como Git, algo que não é tão simples quando trabalhamos apenas com planilhas.

Isso não significa que ferramentas como Excel deixaram de ser importantes. Na verdade, elas continuam sendo muito utilizadas, principalmente para visualizações rápidas e análises exploratórias simples. Porém, quando lidamos com maior volume de dados, automação ou análises mais complexas, o uso de Python e Pandas se torna uma solução muito mais poderosa.

Por esse motivo, o Pandas acabou se tornando uma ferramenta fundamental para profissionais da área de dados, sendo amplamente utilizada em empresas de tecnologia, bancos, startups, consultorias e diversos outros setores.

Durante a aula, iremos explorar na prática como o Pandas pode nos ajudar a ler dados, manipulá-los e extrair informações importantes, mostrando como a programação pode transformar a forma como trabalhamos com dados

Como aplicar filtros?

Nesse caso, podemos usar apenas como uma questão de comparação, mas não da maneira abaixo

```python
df['Idade'] > 30
```

Dessa maneira, ele traz apenas uma Serie Booleana, informando se aquela linha é ou não maior do que 30

```python
df[df['Idade'] > 30]
```

Combinando filtros:

```python
# trazendo todos os funcionários com mais de 30 anos e que ganham mais que 5.000
filtro1 = df['Idade'] > 30
filtro2 = df['Salário'] > 5_000

df[filtro1 & filtro2]
```

Criando novas colunas

```python
df['Salário Anual'] = df['Salário'] * 12
df
```

Também podemos remover as colunas de maneira bem simples

```python
df = df.drop('Salário Anual', axis=1)
```
