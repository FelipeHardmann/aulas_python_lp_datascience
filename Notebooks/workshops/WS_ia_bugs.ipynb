## O que é IA?

Inteligência Artificial (IA) é o ramo da computação que busca criar sistemas capazes de realizar tarefas que exigem inteligência humana, como aprender, raciocinar, reconhecer padrões e se adaptar.

Em outras palavras, podemos dizer que é quando a máquina atinge um nível de inteligência que é capaz de simular a inteligência humana, tomando suas próprias decisões.

### IA na prática

Nessa sessão vamos dividir as práticas de IA no código em algumas partes, a primeira parte que vamos verificar é o debug de um código simples

- 1 - Debug automático:
  Vamos fazer a seguinte atividade agora, eu quero que vocês criem uma função que vai ler uma lista de dicionários, algo semelhante a isso:



```
[
    {"value": 10, "divisor": 2},
    {"value": 20, "divisor": 5},
    {"value": 15, "divisor": 3}
]
```

Nesse caso, eu quero que vocês retornem o resultado da divisão do value pelo divisor.

Vamos lá agora para a resolução desse código, nesse caso, a resolução desse código a grande maioria diria que é a da seguinte forma

````python
def process_data(data):
    results = []
    for item in data:
        processed = item['value'] / item['divisor']
        results.append(processed)
    return results

data = [
    {"value": 10, "divisor": 2},
    {"value": 20, "divisor": 5},
    {"value": 15, "divisor": 3}
]

print(process_data(data))
````

Nesse caso, o código não estaria completamente errado quando olhamos dessa maneira, porém, a IA diria outra coisa, pelo seguinte, existe um caso de erro aí, não tratamos a divisão por 0, então é interessante pedirmos a IA para verificar o que poderia ter errado nessa função

Assim que pedimos a ela para verificar nosso código, de maneira rápida, ela já identifica o erro sem problemas

Então, esse é um caso que podemos usar ela, para debugar o código de maneira rápida.

- Refatoração de código

Vamos fazer o seguinte, temos um código abaixo que retorna se o número é ímpar ou par com base em uma lista



```
def verificar_par_ou_impar(numeros):
    resultado = []
    for i in range(len(numeros)):
        if numeros[i] % 2 == 0:
            resultado.append(str(numeros[i]) + " é par")
        else:
            resultado.append(str(numeros[i]) + " é ímpar")
    return resultado

print(verificar_par_ou_impar([1, 2, 3, 4, 5, 6]))
```

Esse cógido está muito complexo, a atividade que vocês tem agora é tentar melhorar esse código na mão.



O que acontece, mais uma vez a IA consegue nos ajudar bastante nesse tipo de código, principalmente por conta desse código ter a capacidade de ser feito em apenas uma única linha.

Apenas com um CTRL C + CTRL V no input da IA, com alguns dados do que você precisa que seja feito a IA já consegue diminuir esse código.

3 - Otimização de código

Vamos para mais uma atividade, temos o código abaixo, que é bastante lento, pelo seguinte fato, temos dois for dentro desse código, é necessário que seja feita uma refatoração nesse código, para melhorar sua performance, a atividade é tentar fazer essa otimização sem alterar o resultado.

````python
import time

def encontrar_pares_repetidos(lista):
    pares = []
    for i in range(len(lista)):
        for j in range(len(lista)):
            if i != j and lista[i] == lista[j] and lista[i] not in pares:
                pares.append(lista[i])
    return pares

# Teste
numeros = [1, 2, 3, 2, 4, 5, 1, 6, 3]
inicio = time.time()
resultado = encontrar_pares_repetidos(numeros)
fim = time.time()

print("Pares repetidos:", resultado)
print("Tempo de execução:", fim - inicio, "segundos")
````

Bom, esse é o estilo de tarefa que a IA pode te dar dicas ou até mesmo a resposta por completa.

Bom, essas são boas formas de como a IA pode te ajudar no dia a dia, mas lembre-se, é necessário sempre ter conhecimento base daquilo que você trabalha, a IA não fará tudo por você, apenas irá te ajudar no conhecimento.
