### Conceito de Protótipo

Um protótipo é um modelo preliminar ou simulação de um produto, serviço ou sistema, criado para testar e validar ideias, funcionalidades e usabilidade antes do desenvolvimento final.

Nesse sentido, temos diversos exemplos de protótipos, sejam eles no mundo de softwares/sistemas, com aplicativos mobile, sistemas web, sistemas desktops ou até mesmo com serviços onde há a encenação de como aquele serviço vai funcionar e também temos exemplos de produtos físicos com maquetes por exemplo para relacionadas a obras de casas/apartamentos e entre diversos outros.

# O que é o Figma?

## Definição Principal
O **Figma** é uma ferramenta de **design de interface** baseada na **nuvem**, usada principalmente para:

- **Design de UI (Interface do Usuário)**: Criar a parte visual de sites e aplicativos
- **Prototipagem**: Simular como o design vai funcionar, com cliques e transições
- **Collaboração em Tempo Real**: Várias pessoas podem trabalhar no mesmo arquivo simultaneamente

---

## Características Principais

### 🚀 **Funcionalidades Essenciais**
1. **Baseado na Web**
   - Funciona diretamente no navegador
   - Também possui aplicativo para desktop

2. **Collaborativo**
   - Edição em tempo real (estilo Google Docs)
   - Sistema de comentários
   - Compartilhamento facilitado

3. **Multi-plataforma**
   - Compatível com Windows, Mac e Linux

4. **Modelo Freemium**
   - Plano gratuito robusto para indivíduos e pequenos times

5. **Design Systems**
   - Criação de bibliotecas de componentes reutilizáveis
   - Mantém consistência visual entre projetos

---

## Vantagens e Aplicações

### ✅ **Por que é popular?**
- **Elimina arquivos locais**: Não precisa salvar e enviar arquivos manualmente
- **Workflow integrado**: Todo o processo em uma única plataforma
- **Colaboração eficiente**: Time todo trabalha junto no mesmo espaço

### 👥 **Público-alvo**
- **Designers** → Criação de interfaces
- **Desenvolvedores** → Inspeção de código e medidas
- **Product Managers** → Acompanhamento do projeto e feedback
