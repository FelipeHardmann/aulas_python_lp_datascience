Usaremos esses dados para classificar os filmes, visualizar as classifica√ß√µes mai altas, ganhos l√≠quidos, calcular informa√ß√µes estat√≠sticas e etc.

####Primeiros passos

````python
import pandas as pd
````

Come√ßamos importando o pandas ao nosso projeto para utilizarmos as fun√ß√µes que vem com a biblioteca

Em seguida, vamos ler os dados dentro desse arquivo com a fun√ß√£o read_excel.

````python
excel_file = 'movies.xls'
movies = pd.read_excel(excel_file)
````

Atribu√≠mos o read_excel a vari√°vel movies, pois agora essa vari√°vel se torna um DataFrame, que √© um objeto padr√£o do pandas.

````python
print(movies)
````

Podemos visualizar as primeiras linhas de uma maneira melhor, que √© com a fun√ß√£o pr√≥pria do pandas chamada de head, para visualizar as primeiras linhas do nosso DataFrame.

````python
movies.head()
````

Agora, conseguimos ver as primeiras linhas do nosso DataFrame.

Agora, temos uma observa√ß√£o, o arquivo excel normalmente pode ter v√°rias planilhas dentro dele, como √© o caso do nosso, com isso √© interessante que passemos a informa√ß√£o de qual planilha vamos ler esse arquivo, com isso, vamos modificar nossa fun√ß√£o read_excel, que possui mais param√™tros para utilizar.

````python
movies_sheet1 = pd.read_excel(excel_file, sheet_name=0, index_col=0)
movies_sheet1.head()
````

Agora, se quisermos ler a outra planilha, basta modificarmos o sheet_name

````python
movies_sheet2 = pd.read_excel(excel_file, sheet_name=1, index_col=0)
movies_sheet2.head()
````

E a mesma coisa fazemos com a terceira planilha

````python
movies_sheet3 = pd.read_excel(excel_file, sheet_name=2, index_col=0)
movies_sheet3.head()
````

Agora, temos a visualiza√ß√£o de todos os dados das tr√™s planilhas.

Tem uma observa√ß√£o muito interessante nessas tr√™s planilhas, elas possuem as mesmas colunas, o mesmo formato, logo, podemos transformar todas elas em uma √∫nica planilha, da seguinte forma

````python
movies = pd.concat([movies_sheet1, movies_sheet2, movies_sheet3])
````

Com o m√©todo concat, fazemos uma concatena√ß√£o das planilhas, agora temos um DataFrame maior, com todos os dados juntos.

````python
movies.shape
````

Essas s√£o a quantidade de linhas e colunas que temos no nosso DataFrame

Usamos o head() para visualizar as linhas superiores, mas tamb√©m podemos usar o m√©todo tail() para visualizar as inferiores.

````python
movies.tail()
````

No Excel podemos classificar valores de uma planilha se baseando em uma coluna ou mais, no pandas tamb√©m podemos fazer isso da seguinte maneira

````python
sorted_by_gross = movies.sort_values(['Gross Earnings'], ascending=False)
````

````python
sorted_by_gross["Gross Earnings"].head(10)
````

Agora temos, do filme que tem o maior lucro para o menor

Tamb√©m, com a utiliza√ß√£o do Pandas podemos criar gr√°ficos para visualizar a classifica√ß√£o dos filmes que tiveram o maior lucro bruto, para isso, vamos utilizar al√©m do pandas, vamos usar tamb√©m a biblioteca chamada de matplotlib

````python
import matplotlib.pyplot as plt
````

Importado nossa biblioteca, agora vamos utilizar ela para exibir o nosso gr√°fico

````python
sorted_by_gross['Gross Earnings'].head(10).plot(kind="barh")
plt.show()
````

Agora, conseguimos visualizar os filmes com maior lucro bruto de maneira visual atr√°ves do matplot

Agora, vamos criar um histograma de pontua√ß√µes do IMDB para verificar a distribui√ß√£o de pontua√ß√µes do IMDB em todos os filmes. Os histogramas s√£o uma boa maneira de visualizar a distribui√ß√£o de um conjunto de dados.

````python
movies['IMDB Score'].plot(kind="hist")
plt.show()
````

Tamb√©m podemos usar fun√ß√µes para visualizar informa√ß√µes estat√≠sticas da nossa planilha

````python
movies.describe()
````

Mas, em alguns casos n√£o precisamos ver as informa√ß√µes de todas as colunas, para isso, podemos verificar as informa√ß√µes de uma coluna espec√≠fica, utilizando o m√©todo:

Estamos vendo que o Pandas √© muito poderoso para se trabalhar com Planilhas e juntamente com Excel, al√©m dessas quest√µes acima, tamb√©m √© poss√≠vel dividirmos o nosso DataFrame e trabalharmos com subconjuntos de dados.

Da seguinte maneira:

````python
movies_subset_columns = pd.read_excel(excel_file, usecols=[0, 1, 2, 3, 4, 5, 6])
movies_subset_columns.head()
````

Dessa maneira, dividimos o nosso DataFrame, agora, ao inv√©s de utilizarmos as 25 colunas, estamos utilizando apenas 7 delas, isso por conta do param√™tro da fun√ß√£o que √© o usecols, nele informamos os ind√≠ces de cada de colunas que queremos usar, ou seja, um subconjunto de dados.

### Aplicando f√≥rmulas

No Excel temos processos para aplicar f√≥rmulas, para criar novas colunas a partir de algumas existente, podemos criar tabelas din√¢micas, etc.

Com o Python e o Pandas tamb√©m conseguimos aplicar essas modifica√ßoes, da seguinte maneira:

Podemos obter por exemplo, uma nova coluna, dos ganhos l√≠quidos subtraindo os ganhos brutos

````python
movies["Net Earnings"] = movies["Gross Earnings"] - movies["Budget"]
````

````python
movies["Net Earnings"]
````

Que fica dessa maneira que vemos acima, mas, podemos fazer a modifica√ß√£o de visualiza√ß√£o

````python
sorted_movies = movies[['Net Earnings']].sort_values(['Net Earnings'], ascending=[False])
sorted_movies.head(10)['Net Earnings'].plot.barh()
plt.show()
````

Assim, dessa maneira, temos os filmes com o maior lucro l√≠quido, visualmente, atr√°ves de um gr√°fico.

Como foi dito acima, tamb√©m √© poss√≠vel criarmos tabelas din√¢micas, que √© algo at√© um pouco avan√ßado no Excel, primeiro, precisamos identificar quais colunas servir√£o como √≠ndices, vamos por partes, vamos pegar o Ano e Ganhos Brutos

````python
movies_subset = movies[['Year', 'Gross Earnings']]
movies_subset.head()
````

Agora ligamos pivot_tablesobre este subconjunto de dados. O m√©todo pivot_tableleva um par√¢metro index. Como mencionado, queremos usar o Ano como o √≠ndice.

````python
earnings_by_year = movies_subset.pivot_table(index=['Year'])
earnings_by_year.head()
````

Isso acabou de nos dar uma tabela din√¢mica, com o ano como √≠ndice da nossa tabela, agora, a partir dessa tabela din√¢mica podemos usar esses dados para criar uma visualiza√ß√£o gr√°fica, da seguinte maneira

````python
earnings_by_year.plot()
plt.show()
````

Agora, vamos criar outro subconjunto DataFrame, mas desta vez vamos escolher as colunas, Pa√≠s, Idioma e Ganhos Brutos.

````python
movies_subset = movies[['Country', 'Language', 'Gross Earnings']]
movies_subset.head()
````

````python
earnings_by_co_lang = movies_subset.pivot_table(index=['Country', 'Language'])
earnings_by_co_lang.head()
````

Agora, estamos utilizando as colunas pa√≠s e idioma como √≠ndices da nossa tabela din√¢mica, com isso, podemos criar tamb√©m, mais visualiza√ß√µes gr√°ficas

````python
earnings_by_co_lang.head(20).plot(kind='bar', figsize=(20,8))
plt.show()
````

Por fim, com as altera√ß√µes que fizemos, √© sempre interessante que tenhamos o arquivo excel alterado em m√£os, seja para passar para um outro colega ou gestor ou at√© mesmo s√≥ para ter o arquivo, com o Pandas podemos exportar esse arquivo excel da seguinte maneira

````python
movies.to_excel('output.xlsx')
````
